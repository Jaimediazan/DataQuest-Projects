# Using plots to analyze a recent grads dataset

In this project we will use bar, scatter, histograms and box plots to analyze the relationship between several variables related with the job outcomes of students who graduated from college between 2010 and 2012.\
We begin importing data and having a first look on the data.

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline


recent_grads = pd.read_csv('D:/PROYECTOS_PYTHON/recent-grads.csv', encoding = 'UTF8')
recent_grads.iloc[0]
```

```python
print(recent_grads.head())
print('\n')
print(recent_grads.tail())
```

```python
recent_grads.describe()
```

Now, we are going to drop the rows which have some null values in order to not mess up our observations:

```python
raw_data_count = len(recent_grads)
print(raw_data_count)

recent_grads = recent_grads.dropna()
cleaned_data_count = len(recent_grads)
print(cleaned_data_count)
```

In the next step we are going to display some scatter plots to explore the data and compare some variables quickly:

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Median vs. Sample_size', figsize=(8,5))
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Median vs. Sample_size', figsize=(8,5))
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Median vs. Full_time', figsize=(8,5))
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate', figsize=(8,5))
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median', figsize=(8,5))
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median', figsize=(8,5))
```

We make to ourselves four questions:

1) Do students in more popular majors make more money?\
- We can see in the first plot that the most popular majors usually have a median salary below the mean but there's not a strong correlation between two parameters. Besides, in the second plot it seems more obvious than the unemployment rate in the most popular majors is rounding the mean, so we can't say that popularity of the major is important to predict the odds of a situation of unemployment. 

2) Do students that majored in subjects that were majority female make more money?\
- In the fourth plot we check that there are not correlation between the subjects chosen by female and making more money. We can observe how the points in the plot are very scattered, so we can not reach any conclusion about this.

3) Is there any link between the number of full-time employees and median salary?\
- The distribution in the third and the first plot comparing full-time/sample-size with median salary is similar and that's because probably they have similar proportions in all majors. We'll check it in the next cell and then make conclusions.

4) Is there any difference between men and women median salaries?\
- We can observe a slight difference between the majors with a lot of graduated women and graduated men, having in the case of men a better median salary rate.


```python
recent_grads.plot(x='Full_time', y='Sample_size', kind='scatter', title='Full_time vs. Sample_size', figsize=(8,5))
```

As we can see, there is a very strong positive correlation between both variables, so we can not say that there is a link between the number of full-time employees and median salary and we only can make the same conclusions than in the first question.

Other way to do the same faster:

```python
cols = ["Sample_size", "Median", "Employed", "Full_time", "ShareWomen", "Unemployment_rate", "Men", "Women"]
colors = ["blue","red","orange","grey","pink","black","green","purple"]

fig = plt.figure(figsize=(5,30))
for r in range(0,8):
    ax = fig.add_subplot(8,1,r+1)
    ax = recent_grads[cols[r]].plot(kind='hist', rot=30, color=colors[r], title=cols[r])
    plt.subplots_adjust(hspace=0.5)
    
```

Now we are going to join both kind of plots in order to have the whole vision of comparative of two variables.\
For this we'll use scatter matrix:

```python
from pandas.plotting import scatter_matrix
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(15,15))
```

We can confirm that the most of the majors of our data set have less than 500 students so it's comprehensible that these majors have the best and worst median salary rates (there are non-populars major with a high median salary and also with a low median salary).\
In the case of populars major we confirm that are situated near the mean of median salary rates but slightly below (it makes sense because, if there are a lot of people that have studied a major, there will be more candidates in the labor market and the salaries will tend to lower).\
Besides, we can say that a more popular major doesn't mean more unemployment, so we deduce that unemployment rate depends on other factores, but that belongs to other analysis (labor market analysis probably).

### El siguiente paso consistirá en eliminar todas aquellas apps que por su nombre entendemos que no están dirigidas a un público de habla inglesa, excluyendo los registros que contengan caracteres o símbolos no pertenecientes al alfabeto anglosajón.

Para ello, primero realizamos una función que detecte qué apps tienen un nombre con un caracter con un número correspondiente\
mayor a 127 ya que los número 0 a 127 corresponden a los caracteres del alfabeto inglés según el estándar ASCII

```python
def check_english_characters(string):
    
    for character in string:
         if ord(character) > 127:
             return False
    
   
    return True

print(check_english_characters('Instagram'))
print(check_english_characters('爱奇艺PPS -《欢乐颂2》电视剧热播'))
print(check_english_characters('Docs To Go™ Free Office Suite'))
print(check_english_characters('Instachat 😜'))
print ('\n')
print(ord('™'))
print(ord('😜'))
```

La función funciona pero no nos sirve para mantener dentro de nuestra lista de apps algunas con carácteres especiales
que en realidad\
sí están dirigidas al público de habla inglesa (™, 😜). Por ello, modificamos la función para que solo
nos excluya los registros\
con más de 3 caracteres fuera del rango 0-127 previamente marcado:

```python
def check_english_characters(string):
    non_ascii = 0
    for character in string:
         if ord(character) > 127:
             non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:    
        return True


print(check_english_characters('Docs To Go™ Free Office Suite'))
print(check_english_characters('Instachat 😜'))
print(check_english_characters('爱奇艺PPS -《欢乐颂2》电视剧热播'))
```

### A continuación, vamos a ordenar esta información y mostrarla en porcentajes para una visualización más clara

```python
def freq_table(dataset, index): 
    table = {}
    total = 0
    
    for row in dataset:
        total +=1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
            
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total)* 100
        table_percentages[key] = percentage
    
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', round(entry[0],2), '%')
```

Para ello, primero realizamos una funci√≥n que detecte qu√© apps tienen un nombre con un caracter con un n√∫mero correspondiente\
mayor a 127 ya que los n√∫mero 0 a 127 corresponden a los caracteres del alfabeto ingl√©s seg√∫n el est√°ndar ASCII

```python
def check_english_characters(string):
    
    for character in string:
         if ord(character) > 127:
             return False
    
   
    return True

print(check_english_characters('Instagram'))
print(check_english_characters('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
print(check_english_characters('Docs To Go‚Ñ¢ Free Office Suite'))
print(check_english_characters('Instachat üòú'))
print ('\n')
print(ord('‚Ñ¢'))
print(ord('üòú'))
```

La funci√≥n funciona pero no nos sirve para mantener dentro de nuestra lista de apps algunas con car√°cteres especiales
que en realidad\
s√≠ est√°n dirigidas al p√∫blico de habla inglesa (‚Ñ¢, üòú). Por ello, modificamos la funci√≥n para que solo
nos excluya los registros\
con m√°s de 3 caracteres fuera del rango 0-127 previamente marcado:

```python
def check_english_characters(string):
    non_ascii = 0
    for character in string:
         if ord(character) > 127:
             non_ascii += 1
    
    if non_ascii > 3:
        return False
    else:    
        return True


print(check_english_characters('Docs To Go‚Ñ¢ Free Office Suite'))
print(check_english_characters('Instachat üòú'))
print(check_english_characters('Áà±Â•áËâ∫PPS -„ÄäÊ¨¢‰πêÈ¢Ç2„ÄãÁîµËßÜÂâßÁÉ≠Êí≠'))
```

### A continuaci√≥n, vamos a ordenar esta informaci√≥n y mostrarla en porcentajes para una visualizaci√≥n m√°s clara

```python
def freq_table(dataset, index): 
    table = {}
    total = 0
    
    for row in dataset:
        total +=1
        value = row[index]
        if value in table:
            table[value] += 1
        else:
            table[value] = 1
            
    table_percentages = {}
    for key in table:
        percentage = (table[key] / total)* 100
        table_percentages[key] = percentage
    
    return table_percentages

def display_table(dataset, index):
    table = freq_table(dataset, index)
    table_display = []
    for key in table:
        key_val_as_tuple = (table[key], key)
        table_display.append(key_val_as_tuple)
        
    table_sorted = sorted(table_display, reverse = True)
    for entry in table_sorted:
        print(entry[1], ':', round(entry[0],2), '%')
```
